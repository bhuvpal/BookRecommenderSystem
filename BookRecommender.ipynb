```python
x = merge_books_rating.groupby('User-ID').count()['Book-Rating']>200
high_rated_userid = x[x].index
```

```python
high_rated_userid.shape
```

```python
filtered_rating = merge_books_rating[merge_books_rating['User-ID'].isin(high_rated_userid)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
fambooks = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(fambooks)]
```

```python
final_rating.drop_duplicates()
```

```python
final_rating
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_array = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    distances = sorted(list(enumerate(similarity_array[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in distances:
        print(pt.index[i[0]])
```
